# Exploring the Data for House Price

```python
graphlab.canvas.set_target('ipynb')
sales.show()
```

```python
sales.show(view='Scatter Plot', x ='sqft_living',y ='price')
```

# Creating a Simple Linear Regression Model of Sqft_living to Price

```python
training_data,test_data = sales.random_split(0.8,seed=0)
```

```python
training_data.shape
```

```python
test_data.shape
```


# Build the Linear Regression Model

```python
sqft_model = graphlab.linear_regression.create(training_data,target='price',features=['sqft_living'])
```

# Evaluating the Simple Model

```python
test_data['price'].mean()
```

```python
sqft_model.evaluate(test_data)
```

# Let's show what our Predictions look like

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(test_data['sqft_living'],test_data['price'],'.', test_data['sqft_living'],sqft_model.predict(test_data),'-')
```

```python
sqft_model.get('coefficients')
```
